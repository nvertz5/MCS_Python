```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from netCDF4 import Dataset
import cartopy.crs as crs
from cartopy.feature import NaturalEarthFeature
import wrf
from wrf import (to_np, getvar, smooth2d, get_cartopy, cartopy_xlim,
                cartopy_ylim, latlon_coords, interplevel, extract_times)
import metpy.calc as mcalc
from metpy.units import units
```

```python
# Import RUC & WRF netcdf files
ncfile_WRF = Dataset('../../brianjs/MCS_LLJ_study/WRF_3.6.1_3km/wrfout_d01_2011-07-05_1200_UTC_WSM6_YSU.nc')
ncfile_RUC = Dataset('../../brianjs/MCS_LLJ_study/RUC_Analysis/ruc_analysis_07_06_2011.nc')
ncfile_obs = Dataset("../../brianjs/MCS_LLJ_study/obs_reflectivity/refc_07_06_2011.nc")
```

```python
# Create Python variables from RUC/WRF variables
time = 18

# WRF Variables
dbz_WRF = getvar(ncfile_WRF,"dbz",time) # Reflectivity
tempC_WRF = getvar(ncfile_WRF, 'temp', time, units='degC') # Temperature in degC
pres_WRF = getvar(ncfile_WRF, 'pressure', time) # Pressure in hPa
rh_WRF = getvar(ncfile_WRF, 'rh', time) # Relative Humidity as a percentage
lat_WRF, lon_WRF = latlon_coords(pres_WRF) # Latitude & Longitude

# RUC Variables
tempK_RUC = ncfile_RUC.variables['TT'][time,:,:,:] # Temperature in Kelvin
tempC_RUC = tempK_RUC - 273.15                        # Temperature in degC
pres_RUC = ncfile_RUC.variables['PRES'][time,:,:,:] / 100 # Pressure in hPa
rh_RUC = ncfile_RUC.variables['RH'][time,:,:,:] # Relative Humidity as a percentage
lat_RUC = ncfile_RUC.variables['XLAT_M'][time,:,:] # Latitude
lon_RUC = ncfile_RUC.variables['XLONG_M'][time,:,:] # Longitude

# OBS Variables
dbz_OBS = ncfile_obs.variables['REFC_P0_L200_GLL0'][time,:,:] #[time,north-south,east-west
lat_OBS = ncfile_obs.variables['lat_0'][:]
lon_OBS = ncfile_obs.variables['lon_0'][:]

# Create Specific Humidity Variables for WRF/RUC
def rh2qair(rh, temp_C, pres):
    es = 6.112 * np.exp((17.67 * temp_C) / (temp_C + 243.5))
    e = (rh / 100) * es
    qair = (0.622 * e) / (pres - (0.378 * e)) * 1000
    return(qair)

# Create Equivalent Potential Temperature Variables for WRF/RUC
def thetaEcalc(rh, temp_C, pres):
    temp_K = 273.15 + temp_C
    thetaE = temp_K * (1000/pres)**0.286 + (3 * (rh * (3.884266 * 10**((7.5 * temp_C) / (237.7 + temp_C))) / 100))
    return(thetaE)

# Remaining WRF/RUC Variables
qair_WRF = rh2qair(rh_WRF, tempC_WRF, pres_WRF)
qair_RUC = rh2qair(rh_RUC, tempC_RUC, pres_RUC)
thetaE_WRF = thetaEcalc(rh_WRF, tempC_WRF, pres_WRF)
thetaE_RUC = thetaEcalc(rh_RUC, tempC_RUC, pres_RUC)
```

```python
# Create interpolated variables of 925 and 850 mb levels

# 925 mb level
qair925_WRF = interplevel(qair_WRF, pres_WRF, 925)
qair925_RUC = qair_RUC[4,:,:] 
thetaE925_WRF = interplevel(thetaE_WRF, pres_WRF, 925)
thetaE925_RUC = thetaE_RUC[4,:,:]
rh925_WRF = interplevel(rh_WRF, pres_WRF, 925)
rh925_RUC = rh_RUC[4,:,:]

# 850 mb level
qair850_WRF = interplevel(qair_WRF, pres_WRF, 850)
qair850_RUC = qair_RUC[7,:,:] 
thetaE850_WRF = interplevel(thetaE_WRF, pres_WRF, 850)
thetaE850_RUC = thetaE_RUC[7,:,:]
rh850_WRF = interplevel(rh_WRF, pres_WRF, 850)
rh850_RUC = rh_RUC[7,:,:]
```

```python
# Create composite reflectivity for wrf (obs is already composite)
dBZ_WRF = np.zeros(dbz_WRF[0,0].shape)
dBZ_WRF = np.max(dbz_WRF, axis=0)
for i in range(len(dbz_WRF[1,:,1])):
    for j in range(len(dbz_WRF[1,1,:])):
        dBZ_WRF[i,j] = np.max(dbz_WRF[:,i,j])
```

```python
# Smooth WRF Contours
smooth_qair925_WRF = smooth2d(qair925_WRF, 40)
smooth_qair850_WRF = smooth2d(qair850_WRF, 40)
smooth_thetaE925_WRF = smooth2d(thetaE925_WRF, 40)
smooth_thetaE850_WRF = smooth2d(thetaE850_WRF, 40)
smooth_rh925_WRF = smooth2d(rh925_WRF, 40)
smooth_rh850_WRF = smooth2d(rh850_WRF, 40)

smooth_qair925_RUC = smooth2d(qair925_RUC, 40)
smooth_qair850_RUC = smooth2d(qair850_RUC, 40)
smooth_thetaE925_RUC = smooth2d(thetaE925_RUC, 40)
smooth_thetaE850_RUC = smooth2d(thetaE850_RUC, 40)
smooth_rh925_RUC = smooth2d(rh925_RUC, 40)
smooth_rh850_RUC = smooth2d(rh850_RUC, 40)

# Create Figure
fig = plt.figure(figsize=(20,15))
ax_RUCrh = fig.add_subplot(221,projection=crs.LambertConformal())
ax_WRFrh = fig.add_subplot(222,projection=crs.LambertConformal())
ax_RUCthetaE = fig.add_subplot(223,projection=crs.LambertConformal())
ax_WRFthetaE = fig.add_subplot(224,projection=crs.LambertConformal())

# Set GeoAxes
#ax_RUCrh = plt.axes(projection=crs.LambertConformal())
ax_RUCrh.set_extent([-87.5,-104.,32.,45.75])
#ax_WRFrh = plt.axes(projection=crs.LambertConformal())
ax_WRFrh.set_extent([-87.5,-104.,32.,45.75])
#ax_RUCthetaE = plt.axes(projection=crs.LambertConformal())
ax_RUCthetaE.set_extent([-87.5,-104.,32.,45.75])
#ax_WRFthetaE = plt.axes(projection=crs.LambertConformal())
ax_WRFthetaE.set_extent([-87.5,-104.,32.,45.75])
```

```python
# Add state boundaries & coastlines to plot
states_provinces = NaturalEarthFeature(
    category='cultural',
    name='admin_1_states_provinces_lines',
    scale='50m',
    facecolor='none')

ax_RUCrh.add_feature(states_provinces, edgecolor='black', linewidth=0.5)
ax_RUCrh.coastlines('50m',linewidth=0.8)
ax_WRFrh.add_feature(states_provinces, edgecolor='black', linewidth=0.5)
ax_WRFrh.coastlines('50m',linewidth=0.8)
ax_RUCthetaE.add_feature(states_provinces, edgecolor='black', linewidth=0.5)
ax_RUCthetaE.coastlines('50m',linewidth=0.8)
ax_WRFthetaE.add_feature(states_provinces, edgecolor='black', linewidth=0.5)
ax_WRFthetaE.coastlines('50m',linewidth=0.8)
```

```python
# Plot filled contours for sh at 925 hPa
levels_qair = [8,10,12,14,16,18,20]

cs1_RUCrh = ax_RUCrh.contour(lon_RUC,lat_RUC,to_np(rh850_RUC), 
                            transform=crs.PlateCarree(), colors='tab:red')
cs2_RUCrh = ax_RUCrh.contour(lon_OBS, lat_OBS, to_np(dbz_OBS), levels = [40],
                             transform=crs.PlateCarree(), colors='tab:blue')
cs3_RUCrh = ax_RUCrh.contourf(lon_RUC,lat_RUC,to_np(qair850_RUC),
                             levels_qair,
                             transform=crs.PlateCarree(),cmap=get_cmap('Greens'))
ax_RUCrh.grid(True)

cs1_WRFrh = ax_WRFrh.contour(lon_WRF,lat_WRF,to_np(smooth_rh850_WRF), 
                            transform=crs.PlateCarree(), colors='tab:red')
cs2_WRFrh = ax_WRFrh.contour(lon_OBS, lat_OBS, to_np(dbz_OBS), levels = [40],
                             transform=crs.PlateCarree(), colors='tab:blue')
cs3_WRFrh = ax_WRFrh.contourf(lon_WRF,lat_WRF,to_np(smooth_qair850_WRF),
                             levels_qair,
                             transform=crs.PlateCarree(),cmap=get_cmap('Greens'))
ax_WRFrh.grid(True)

cs1_RUCthetaE = ax_RUCthetaE.contour(lon_RUC,lat_RUC,to_np(thetaE850_RUC), 
                            transform=crs.PlateCarree(), colors='tab:red')
cs2_RUCthetaE = ax_RUCthetaE.contour(lon_OBS, lat_OBS, to_np(dbz_OBS), levels = [40],
                             transform=crs.PlateCarree(), colors='tab:blue')
cs3_RUCthetaE = ax_RUCthetaE.contourf(lon_RUC,lat_RUC,to_np(qair850_RUC),
                             levels_qair,
                             transform=crs.PlateCarree(),cmap=get_cmap('Greens'))
ax_RUCrh.grid(True)

cs1_WRFthetaE = ax_WRFthetaE.contour(lon_WRF,lat_WRF,to_np(smooth_thetaE850_WRF), 
                            transform=crs.PlateCarree(), colors='tab:red')
cs2_WRFthetaE = ax_WRFthetaE.contour(lon_OBS, lat_OBS, to_np(dbz_OBS), levels = [40],
                             transform=crs.PlateCarree(), colors='tab:blue')
cs3_WRFthetaE = ax_WRFthetaE.contourf(lon_WRF,lat_WRF,to_np(smooth_qair850_WRF),
                             levels_qair,
                             transform=crs.PlateCarree(),cmap=get_cmap('Greens'))
ax_WRFrh.grid(True)

# Create labeling stuff
lines_RUCrh = [ cs1_RUCrh.collections[0], cs2_RUCrh.collections[-1]]
labels_RUCrh = ['RH 925 hPa - RUC','Obs Composite Reflectivity > 40 dbz']
ax_RUCrh.legend(lines_RUCrh, labels_RUCrh, loc='upper left')
ax_RUCrh.set_title('RUC Analysis with Relative Humidity')
plt.clabel(cs1_RUCrh, inline=1, font=10)
plt.colorbar(cs3_RUCrh, ax=ax_RUCrh, shrink=0.62)

lines_WRFrh = [ cs1_WRFrh.collections[0], cs2_WRFrh.collections[-1]]
labels_WRFrh = ['RH 925 hPa - WRF','Obs Composite Reflectivity > 40 dbz']
ax_WRFrh.legend(lines_WRFrh, labels_WRFrh, loc='upper left')
ax_WRFrh.set_title('WRF Analysis with Relative Humidity')
plt.clabel(cs1_WRFrh, inline=1, font=10)
plt.colorbar(cs3_WRFrh, ax=ax_WRFrh, shrink=0.62)

lines_RUCthetaE = [ cs1_RUCthetaE.collections[0], cs2_RUCthetaE.collections[-1]]
labels_RUCthetaE = ['thetaE 925 hPa - RUC','Obs Composite Reflectivity > 40 dbz']
ax_RUCthetaE.legend(lines_RUCthetaE, labels_RUCthetaE, loc='upper left')
ax_RUCthetaE.set_title('RUC Analysis with Equivalent Pot. Temp.')
plt.clabel(cs1_RUCthetaE, inline=1, font=10)
plt.colorbar(cs3_RUCthetaE, ax=ax_RUCthetaE, shrink=0.62)

lines_WRFthetaE = [ cs1_WRFthetaE.collections[0], cs2_WRFthetaE.collections[-1]]
labels_WRFthetaE = ['thetaE 925 hPa - WRF','Obs Composite Reflectivity > 40 dbz']
ax_WRFthetaE.legend(lines_WRFthetaE, labels_WRFthetaE, loc='upper left')
ax_WRFthetaE.set_title('WRF Analysis with Equivalent Pot. Temp.')
plt.clabel(cs1_WRFthetaE, inline=1, font=10)
plt.colorbar(cs3_WRFthetaE, ax=ax_WRFthetaE, shrink=0.62)

plt.show()
```
