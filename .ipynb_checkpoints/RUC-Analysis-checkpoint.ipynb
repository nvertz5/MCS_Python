```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from netCDF4 import Dataset
import cartopy.crs as crs
from cartopy.feature import NaturalEarthFeature
import wrf
from wrf import (to_np, getvar, smooth2d, get_cartopy, cartopy_xlim,
                cartopy_ylim, latlon_coords, interplevel, extract_times)
```

```python
#Load RUC Data
ncfile_RUC = Dataset('../../brianjs/MCS_LLJ_study/RUC_Analysis/ruc_analysis_08_23_2015.nc')
ncfile_WRF = Dataset('../../brianjs/MCS_LLJ_study/WRF_3.6.1_3km/wrfout_d01_2015-08-22_1200_UTC_WSM6_YSU.nc')
```

```python
#print(ncfile_RUC.variables['PRES'][0,:,0,0])
```

```python
#Grab python variables from netcdf(RUC) variables
time = 15

# RUC Variables
tempK_RUC = ncfile_RUC.variables['TT'][time,:,:,:] # Temperature in Kelvin
tempC_RUC = tempK_RUC - 273.15                        # Temperature in degC
pres_RUC = ncfile_RUC.variables['PRES'][time,:,:,:] / 100 # Pressure in hPa
rh_RUC = ncfile_RUC.variables['RH'][time,:,:,:] # Relative Humidity as a percentage
lat_RUC = ncfile_RUC.variables['XLAT_M'][time,:,:] # Latitude
lon_RUC = ncfile_RUC.variables['XLONG_M'][time,:,:] # Longitude

# WRF Variables
tempC_WRF = getvar(ncfile_WRF, 'temp', time, units='degC') # Temperature in degC
pres_WRF = getvar(ncfile_WRF, 'pressure', time) # Pressure in hPa
rh_WRF = getvar(ncfile_WRF, 'rh', time) # Relative Humidity as a percentage
lat_WRF, lon_WRF = latlon_coords(pres_WRF) # Latitude & Longitude

# Create Specific Humidity variable
def rh2qair(rh, temp_C, pres):
    es = 6.112 * np.exp((17.67 * temp_C) / (temp_C + 243.5))
    e = (rh / 100) * es
    qair = (0.622 * e) / (pres - (0.378 * e)) * 1000
    return(qair)

qair_RUC = rh2qair(rh_RUC, tempC_RUC, pres_RUC) # Specific Humidity in g/kg
qair900_RUC = qair_RUC[5,:,:]           # Specific
qair_WRF = rh2qair(rh_WRF, tempC_WRF, pres_WRF)
qair900_WRF = interplevel(qair_WRF, pres_WRF, 900)
```

```python
# Smooth QAir WRF Contours
smooth_qair900_WRF = smooth2d(qair900_WRF, 10)

# Create Figure
fig = plt.figure(figsize=(14,11))

# Set GeoAxes
ax = plt.axes(projection=crs.LambertConformal())
ax.set_extent([-87.5,-104.,32.,45.75])
```

```python
# Add state boundaries & coastlines to plot
states_provinces = NaturalEarthFeature(
    category='cultural',
    name='admin_1_states_provinces_lines',
    scale='50m',
    facecolor='none')
ax.add_feature(states_provinces, edgecolor='black', linewidth=0.5)
ax.coastlines('50m',linewidth=0.8)
```

```python
# Plot filled contours for sh at 900 hPa
levels_qair = [8,10,12,14,16,18,20]
cs_RUC = ax.contour(lon_RUC,lat_RUC,to_np(qair900_RUC), levels_qair, 
                  transform=crs.PlateCarree(), colors='tab:red')

cs_WRF = ax.contour(lon_WRF, lat_WRF, to_np(smooth_qair900_WRF), levels_qair,
                    transform=crs.PlateCarree(), colors='tab:blue')

# Create legend stuff
lines = [ cs_RUC.collections[0], cs_WRF.collections[-1]]
labels = ['QAir 900 hPa - RUC','QAir 900 hPa - WRF']
ax.legend(lines, labels, loc='upper left')
plt.clabel(cs_RUC, inline=1, font=10)
plt.clabel(cs_WRF, inline=1, font=10)

fig
```
