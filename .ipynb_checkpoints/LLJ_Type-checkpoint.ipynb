```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from netCDF4 import Dataset
import cartopy.crs as crs
from cartopy.feature import NaturalEarthFeature
import wrf
from wrf import (to_np, getvar, smooth2d, get_cartopy, cartopy_xlim,
                cartopy_ylim, latlon_coords, interplevel, extract_times)
import metpy.calc as mcalc
from metpy.units import units
```

```python
# Import RUC File
ncfile_RUC = Dataset('../../brianjs/MCS_LLJ_study/RUC_Analysis/ruc_analysis_07_06_2011.nc')
ncfile_OBS = Dataset('../../brianjs/MCS_LLJ_study/obs_reflectivity/refc_07_06_2011.nc')
```

```python
# Create python variables from ruc variables
time = 18

# RUC Variables
u_wind = ncfile_RUC.variables['UU'][time,:,:,:] * units('m/s') # Zonal wind component in m/s
v_wind = ncfile_RUC.variables['VV'][time,:,:,:] * units('m/s')# Meridional wind component in m/s
lat_RUC = ncfile_RUC.variables['XLAT_M'][time,:,:] # Latitude
lon_RUC = ncfile_RUC.variables['XLONG_M'][time,:,:] # Longitude

# OBS Variables
dbz_OBS = ncfile_OBS.variables['REFC_P0_L200_GLL0'][time,:,:] #[time,north-south,east-west
lat_OBS = ncfile_OBS.variables['lat_0'][:]
lon_OBS = ncfile_OBS.variables['lon_0'][:]

# Destagger u,v components
u_wind_destagger = 0.5 * (u_wind[:,:,:-1] + u_wind[:,:,:-1])
v_wind_destagger = 0.5 * (v_wind[:,:-1,:] + v_wind[:,:-1,:])

# Create 900 & 200 mb level u/v wind components
u_wind900 = u_wind_destagger[5,:,:]
v_wind900 = v_wind_destagger[5,:,:]
u_wind200 = u_wind_destagger[33,:,:]
v_wind200 = v_wind_destagger[33,:,:]

# Create 900 & 200 mb wind speeds
wspd900 = (u_wind900 ** 2 + v_wind900 ** 2) ** 0.5
wspd200 = (v_wind200 ** 2 + v_wind200 ** 2) ** 0.5

#Create np.ndarray variables for 900 & 200 mb u,v components
u_wind900_2 = np.asarray(u_wind900)
v_wind900_2 = np.asarray(v_wind900)
u_wind200_2 = np.asarray(u_wind200)
v_wind200_2 = np.asarray(v_wind200)
```

```python
# Create 900 mb convergence & 900 mb divergence variables 
conv900 = mcalc.h_convergence(u_wind900, v_wind900, 13000 * units('m'), 13000 * units('m'))
div200 = mcalc.h_convergence(u_wind200, v_wind200, 13000 * units('m'), 13000 * units('m'))

# Make the values 10 ** -5
conv900 = conv900 * 10 ** 5
div200 = div200 * 10 ** 5
```

```python
# Create figure and subplots with axes extent
fig = plt.figure(figsize=(20,15))
ax_900 = fig.add_subplot(221, projection=crs.LambertConformal())
ax_200 = fig.add_subplot(222, projection=crs.LambertConformal())
ax_conv = fig.add_subplot(212, projection=crs.LambertConformal())

ax_900.set_extent([-87.5,-104.,32.,45.75])
ax_200.set_extent([-87.5,-104.,32.,45.75])
ax_conv.set_extent([-87.5,-104.,32.,45.75])
```

```python
# Add state boundaries & coastlines to plot
states_provinces = NaturalEarthFeature(
    category='cultural',
    name='admin_1_states_provinces_lines',
    scale='50m',
    facecolor='none')

ax_900.add_feature(states_provinces, edgecolor='black', linewidth=0.5)
ax_900.coastlines('50m',linewidth=0.8)
ax_200.add_feature(states_provinces, edgecolor='black', linewidth=0.5)
ax_200.coastlines('50m',linewidth=0.8)
ax_conv.add_feature(states_provinces, edgecolor='black', linewidth=0.5)
ax_conv.coastlines('50m',linewidth=0.8)
```

```python
# Make dem plots yo
levels_900 = [10,15,100]
levels_200 = [15,30,100]
levels_conv = [-10,-7.5,-5,]
levels_div = [5,7.5,10]
levels_dbz = [45,100]

cs1_900 = ax_900.contourf(lon_RUC, lat_RUC, wspd900,levels_900,
                          transform=crs.PlateCarree(),
                          cmap=get_cmap("YlOrRd"))
cs2_900 = ax_900.streamplot(lon_RUC, lat_RUC, u_wind900_2, v_wind900_2,
                            transform=crs.PlateCarree())
ax_900.grid(True)

cs1_200 = ax_200.contourf(lon_RUC, lat_RUC, wspd200, levels_200,
                         transform=crs.PlateCarree(),
                         cmap=get_cmap("YlOrRd"))
cs2_200 = ax_200.streamplot(lon_RUC, lat_RUC, u_wind200_2, v_wind200_2,
                           transform=crs.PlateCarree())
ax_200.grid(True)

cs_conv = ax_conv.contour(lon_RUC, lat_RUC, conv900, levels_conv,
                          transform=crs.PlateCarree(),
                          colors='tab:red')
cs_div = ax_conv.contour(lon_RUC, lat_RUC, div200, levels_div,
                         transform=crs.PlateCarree(),
                         colors='tab:blue')
cs_refl = ax_conv.contourf(lon_OBS, lat_OBS, dbz_OBS, levels_dbz,
                         transform=crs.PlateCarree(),
                         cmap=get_cmap('Greens'))
ax_conv.grid(True)

# Create labeling stuff
#lines_900 = [ cs2_900.collections[0]]
#labels_900 = ['900 mb streamlines']
#ax_900.legend(lines_900, labels_900, loc='upper left')
ax_900.set_title('900 mb analysis of streamlines and wind magnitudes')
ax_200.set_title('200 mb analysis of streamlines and wind magnitudes')
ax_conv.set_title('900 mb Convergence in red (dashed) & 200 mb Divergence in blue (solid)')
plt.clabel(cs_conv, inline=1, font=10)
plt.clabel(cs_div, inline=1, font=10)
plt.colorbar(cs1_900, ax=ax_900, shrink=0.62)
plt.colorbar(cs1_200, ax=ax_200, shrink=0.62)

plt.show()
```
